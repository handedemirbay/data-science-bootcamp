Makine öğrenimi modelleri için sayısal olmayan değerleri sayısal değerlere dönüştürmemiz gerekiyor. Bunun için iki seçeneğimiz var; Label Encoder ve Pandas get_dummies yöntemi. get_dummies kullanıyoruz.

```python
dummies = ['fare', 'age', 'embarked', 'sex']
dummy_df = pd.get_dummies(df_4[dummies])
```

'dummies' sayısal değerlere dönüştürmek istediğimiz sütun adlarını içerir. Her değişken bir sütun olacak ve 0 ile 1 olarak tanımlanacak.

```python
dummy_df
```

```python
df_5 = pd.concat([df_4, dummy_df], axis = 1)
df_5.drop(dummies, axis=1, inplace=True)
```

```python
df_5
```

```python
# İki veri çerçevesini birleştireceğiz ve eski sütunları bırakacağız

df_5 = pd.concat([df_4, dummy_df], axis = 1)
df_5.drop(dummies, axis=1, inplace=True)
```

```python
df_5
```

```python
df_5.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
X = df_5.drop('survived', axis = 1)
y = df_5['survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
y_pred = log_reg.predict(X_test)
y_pred
```

```python
accuracy_score(y_pred, y_test)
```

```python
confusion_matrix(y_pred, y_test)
```
