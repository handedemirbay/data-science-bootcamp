```python
import matplotlib.pyplot as plt
import seaborn as sns

class DataOperations:
    def __init__(self):
        self.data = sns.load_dataset("penguins")

    def describe(self):
        print(self.data.describe())

    def visulation(self):
    
        '''Penguen veri seti g√∂rselleri'''
        plt.figure()
        sns.countplot(x="body_mass_g", data=self.data)


        plt.figure()
        sns.stripplot(x="species", y="bill_length_mm", data=self.data)

        plt.figure()
        g = sns.PairGrid(self.data, hue="species")
        g.map_diag(sns.histplot)
        g.map_offdiag(sns.scatterplot)


sample = DataOperations()
sample.describe()
sample.visulation()
print(sample.visulation.__doc__)
```
